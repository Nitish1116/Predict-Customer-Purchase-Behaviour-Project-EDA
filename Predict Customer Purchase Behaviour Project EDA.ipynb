```python
#Loyalty Program Members vs Purchase Status
plt.figure(figsize=(8,6))
sns.countplot(data,x='LoyaltyProgram', hue='PurchaseStatus', palette='pastel')
plt.title('Loyalty Program Members vs Purchase Status')
plt.xlabel('Loyalty Program (0: No, 1: Yes)')
plt.ylabel('Count')
plt.legend(title='Purchase Status')
plt.show()
```

```python
avg_purchases = data.groupby('ProductCategory')['NumberOfPurchases'].mean().sort_values()
plt.figure(figsize=(12, 8))
avg_purchases.plot(kind='bar', color='skyblue')
plt.xlabel('Product Category')
plt.ylabel('Average Number of Purchases')
plt.title('Average Number of Purchases by Product Category')
plt.xticks(rotation=45)
plt.show()
```

```python
status_counts = data['PurchaseStatus'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Purchase Status')
plt.show()
```
